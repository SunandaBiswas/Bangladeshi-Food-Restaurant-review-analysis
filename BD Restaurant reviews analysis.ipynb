```python
import nltk
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
df = pd.read_csv(r'C:\Users\sunan\Downloads\reviews.csv')
df
```

```python
df.isnull().sum()
```

```python
df.isna().sum().any
```

```python
df = df[['Review Text', 'Review' ]]
df
```

```python
df['Review Text'] = df['Review Text'].fillna(0)
df['Review'] = df['Review'].fillna(0)
df
```

```python
for i, row in df.iterrows():
    if df['Review'][i] == 5:
        df['Review'][i] = 'Highly Positive'
    elif df['Review'][i] ==4:
        df['Review'][i] = 'Positive' 
    elif df['Review'][i] == 3:
         df['Review'][i] = 'Moderate' 
    elif df['Review'][i] >= 1 or df['Review'][i] <=2:
        df['Review'][i] = 'Negative' 
    elif df['Review'][i] == 0:
        df['Review'][i] = 'Neutral' 
```

```python
# Cleaning the texts
corpus = []
for i in range(0,16597):
    review = re.sub('[^a-zA-Z]', ' ',df['Review Text'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#  Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 17000)
X = cv.fit_transform(corpus).toarray()
y = df['Review'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 10)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,y_pred))
```

```python
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(X_train, y_train)
y_pred = MNB.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,y_pred))
```

```python
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(X_train, y_train)
y_pred = GNB.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,y_pred))
```

```python
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
kc = KNeighborsClassifier(n_neighbors=5)
kc.fit(X_train, y_train)
y_pred = kc.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,y_pred))
```

```python
print(metrics.accuracy_score(y_test,y_pred))
```
